```python
numerical_features = ['age','height', 'weight',
       'overallRank', 'overall_score_2023', 'score_reps_2023_1',
       'score_reps_2023_2', 'score_reps_2023_3', 'rank_2023_1', 'rank_2023_2',
       'rank_2023_3', 'rank_2023_4', 'overall_score_2022', 'score_reps_2022_1',
       'score_reps_2022_2', 'score_reps_2022_3', 'overall_score_2021',
       'score_reps_2021_1', 'score_reps_2021_2', 'score_reps_2021_3',
       'score_reps_2021_4', 'overall_score_2020', 'score_reps_2020_2',
       'score_reps_2023_4', 'score_time_2023_4', 'score_reps_2020_1',
       'score_time_2020_1', 'score_reps_2020_3', 'score_time_2020_3',
       'score_reps_2020_4', 'score_time_2020_4', 'score_reps_2020_5',
       'score_time_2020_5']

corr = df[numerical_features].corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))
 
f, ax = plt.subplots(figsize=(11, 9))
 
cmap = sns.diverging_palette(230, 20, as_cmap=True)
 
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
